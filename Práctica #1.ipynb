TRANSFORMACIONES

```python
df_dropped = df.dropna(subset=['count'])
df_dropped.head()
```

```python
df_dropped.shape
```

```python
df_copy = df.copy() #COPIA
df_copy.head()
```

```python
df_dropped['count'].dtype  #TIPO DE VARIABLE
```

```python
df_dropped['count'].mean() #MEDIA
```

```python
df_copy['count'].fillna(value=np.round(df['count'].mean(),decimals=0), inplace=True)
```

```python
df_copy.agg(['count', 'size', 'nunique'])
```

```python
df_sample = df_copy.sample(frac=0.05, random_state=1)
df_sample.head()
```

Normalizar

```python
from sklearn import preprocessing

df_normalized = df.dropna().copy()
min_max_scaler = preprocessing.MinMaxScaler()
np_scaled = min_max_scaler.fit_transform(df_normalized['population'].values.reshape(-1,1))
df_normalized['population_norm'] = np_scaled.reshape(-1,1)
df_normalized.head()
```

```python
#VALORES NORMALES
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# EJES
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# DENSIDAD PUNTOS
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

# HISTOGRAMAS EN EJES
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```
